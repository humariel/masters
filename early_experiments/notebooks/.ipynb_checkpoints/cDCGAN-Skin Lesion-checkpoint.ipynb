```python
def show_plot(examples, n):
    # plot images
    for i in range(n * n):
    # define subplot
        pyplot.subplot(n, n, 1 + i)
        # turn off axis
        pyplot.axis('off')
        # plot raw pixel data
        pyplot.imshow(examples[i, :, :, :])
    pyplot.show()
```

```python
#FAKE SAMPLES OF CLASS 0 (MEL)
g_model = load_model('../history/cDCGAN/training_checkpoints/generator-e100.h5')
latent_points, labels = generate_latent_points(100, 9)
labels = asarray([0 for _ in range(9)])
X = g_model.predict([latent_points, labels])
X = (X + 1) / 2.0
show_plot(X, 3)
```
