```python
# Train model and save loss history
dataset, test_images, test_labels = util.get_mnist()
acgan = ACGAN(epochs=50, batch_sz=256, n_classes=10)
d_fake_losses, d_real_losses, g_losses = acgan.train(dataset)
_eval = acgan.discriminator.evaluate(test_images, [np.ones(10000,), test_labels])
```

```python
d_fake_loss = [x[1] for x in d_fake_losses]
d_fake_cat_loss = [x[2] for x in d_fake_losses]

d_real_cat_loss = [x[2] for x in d_real_losses]

g_loss = [x[1] for x in d_fake_losses]

plt.plot(d_fake_loss, label="discriminator binary crossentropy")
plt.plot(g_loss, label="generator biunary crossentropy")
plt.plot(d_real_cat_loss, label="discriminator real categorical crossentropy")
plt.plot(d_fake_cat_loss, label="discriminator fake categorical crossentropy")
plt.legend(loc='upper right')
```
